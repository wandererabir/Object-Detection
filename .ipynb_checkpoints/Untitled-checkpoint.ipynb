```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
configfile='ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
fmodel='frozen_inference_graph.pb'
```

```python
model=cv2.dnn_DetectionModel(fmodel,configfile)
```

```python
Labels=[]
filename='labels.txt'
with open(filename,'rt') as file:
    Labels=file.read().rstrip('\n').split('\n')
```

```python
print(Labels)
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

```python
img=cv2.imread('testimg1.jpg')
```

```python
plt.figure(figsize = (90,2))
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
ClassIndex, confidece, bbox = model.detect(img,confThreshold=0.5)
print(ClassIndex)

fscale =3
font=cv2.FONT_HERSHEY_COMPLEX_SMALL
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
    cv2.rectangle(img, boxes, (255,0,0),2)
    cv2.putText(img, Labels[ClassInd-1],(boxes[0]+10, boxes[1]+40), font, fontScale=fscale, color=(0,255,0), thickness=3)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
